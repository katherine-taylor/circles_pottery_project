## Diameter finding process
1) Pick four points on the circle
2) Connect two points to create two chords (need slope of the two chords)
3) Construct the perpendicular bisector of the chord (midpoint and perpendicular line)
4) Find the intersection of the two perpendicular bisectors (solution to system of equations)
5) Using center of circle and a point, use the distance formula to find the radius
6) Double to get the diameter
